```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# x = [4,5,10,4,3,11,14,6,10,12]
# y = [21,19,24,17,16,25,24,22,21,21]
x = [1,2,4,5]
y = [1,1,3,4]

data = list(zip(x,y))
print(data)

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
plt.scatter(x,y, c=kmeans.labels_)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('/content/gender_submission (1).csv')
print(data)

selected_columns = ['PassengerId', 'Survived']
combined_data = data[selected_columns]

combine = list(zip(combined_data['PassengerId'], combined_data['Survived']))


kmeans = KMeans(n_clusters=3)
kmeans.fit(combine)
clusters = kmeans.labels_


plt.scatter(combined_data['PassengerId'], combined_data['Survived'], c=clusters)
plt.xlabel('PassengerId')
plt.ylabel('Survived')
plt.title('KMeans Clustering')
plt.show()
```
