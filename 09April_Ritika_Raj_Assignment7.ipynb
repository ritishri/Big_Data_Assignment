```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [2,4,6,8]
y = [3,7,5,10]
sum1 = sum(x)
sum2 = sum(y)
# print(sum1)
# print(sum2)

a = 0
for i in x:
  a += i*i
# print(a)

product = 0
for i in range(len(x)):
  product += x[i] * y[i]
# print(product)


b = len(x) * product - sum1 * sum2
b1 = len(x) * a - (sum1 * sum1)

total = b/b1
print(total)

a = sum2 * a - sum1 * product
total1 = a/b1
print(total1)
print("Linear regression equation: y =", total1, "+", total, "x")

regression_line = [total1 + total * xi for xi in x]
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, regression_line, color='red', label='Linear Regression')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()


plt.grid(True)
plt.show()
```
